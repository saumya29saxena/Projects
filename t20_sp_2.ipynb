```python
import pickle

with open('dataset_level2.pkl', 'rb') as f:
  data = pickle.load(f)
  print(data)
```

```python
df=data
```

```python
from IPython.display import display
display(df)
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
import numpy as np
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df
```

```python
import pandas as pd
```

```python
def convert_player_dismissed(value):
    if value == '0':
        return 0
    else:
        return 1

df['player_dismissed'] = df['player_dismissed'].apply(convert_player_dismissed)

# Step 3: Ensure the 'player_dismissed' column is of integer type
df.loc[:, 'player_dismissed'] = df['player_dismissed'].astype('int')

# Step 4: Calculate the cumulative dismissals per match
df['cumulative_dismissals'] = df.groupby('match_id')['player_dismissed'].transform(pd.Series.cumsum)

# Step 5: Calculate the number of wickets left
df.loc[:, 'wickets_left'] = 10 - df['cumulative_dismissals']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
df.drop(columns=['cumulative_dismissals'],inplace=True)
```

```python
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')

# Fill NaN values in the 'runs' column with 0
df['runs'] = df['runs'].fillna(0)

# Now proceed with the groupby and rolling sum calculation
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    group = groups.get_group(id)
    rolling_sum = group['runs'].rolling(window=30, min_periods=5).sum()
    last_five.extend(rolling_sum.tolist())

print(last_five)
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df.rename(columns={'bowling team': 'bowling_team'}, inplace=True)
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
